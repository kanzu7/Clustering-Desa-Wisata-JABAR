# KLAKTERISASI POTENSI DESA WISATA DI JAWA BARAT PADA TAHUN 2020

Klakterisasi di bagi menjadi 3 yaitu: <br>
1. Berpotensi tinggi
2. Berpotensi sedang
3. Berpotensi rendah

Klakterisasi dilakukan berdasarkan beberapa aspek penilaian pada ADWI 2022 yang datanya dimiliki dan juga ditambahkan satu aspek baru berdasarkan data yang dimiliki saat ini.

## 7 ASPEK PENILAIAN PADA ADWI 2022

1. DAYA TARIK PENGUNJUNG => Memiliki keunikan & keotentikan produk wisata yang bervariatif dan kreatif
    
2. HOMESTAY => Peningkatan standar kualitas pelayanan serta melestarikan disain arsitektur budaya lokal
    
3. DIGITAL DAN KREATIF => Akselerasi percepatan transformasi digital serta menciptakan konten-konten kreatif
    
4. SUVENIR => Menggali potensi kreativitas dan hasil karya masyarakat berbasis kearifal lokal
    
5. TOILET UMUM => Upaya pemenuhan sarana & prasarana demi kenyamanan wisatawan
    
6. CHSE => Penerapan fasilitas CHSE mengikuti standar nasional
    
7. KELEMBAGAAN DESA => Terbentuknya legalitas berbadan hukum dan pengelolaan yang berkelanjutan

Dari data yang digunakan tersebut akan dilakukan scoring untuk mendapatkan desa yang memiliki kelembagaan desa yang memadai, cukup memadai, dan kurang memadai.

Adapun scoring yang dilakukan sebagai berikut:
- Desa yang memiliki nilai satuan 0 akan mendapatkan 3 poin
- Desa yang memiliki nilai satuan 1 akan mendapatkan 0 poin <br>
<br>
- Desa yang memiliki nilai satuan 1 akan mendapatkan 3 poin
- Desa yang memiliki nilai satuan 2 akan mendapatkan 2 poin
- Desa yang memiliki nilai satuan 3 akan mendapatkan 0 poin

Keseluruhan poin akan dijumlahkan sehingga setiap desa akan mendapatkan total score masing-masing.<br>
Dari nilai total score tersebut akan dihitung indeks kelayakan yang akan digunakan untuk menentukan kelas setiap desa.

```python
aspekDesaWisata = df[['kemendagri_kode_kecamatan', 'kemendagri_nama_kecamatan', 'kemendagri_kode_desa_kelurahan',
          'kemendagri_nama_desa_kelurahan', 'ekonomi_hotel','ekonomi_produk_unggul','ekologi_dampak_cemar',
          'ekonomi_status_bumdesa','ekonomi_bumdes_bid_pariwisata', 'ekonomi_pariwisata_wisata_desa',
          'ekonomi_pariwisata_agrowisata','ekonomi_pariwisata_wisata_alam','ekonomi_pariwisata_transportasi',
          'ekologi_airdesa', 'ekonomi_transumum', 'ekonomi_jalan_desa', 'ekonomi_kualitas_jalan',
          'sosial_bidang_pemukiman_sinyal']]
aspekDesaWisata.head(2)
```

```python
aspekDesaWisata['ekonomi_jalan_desa'] = aspekDesaWisata['ekonomi_jalan_desa'].replace(0, 3)
aspekDesaWisata['sosial_bidang_pemukiman_sinyal'] = aspekDesaWisata['sosial_bidang_pemukiman_sinyal'].replace(0, 3)
aspekDesaWisata['ekologi_dampak_cemar'] = aspekDesaWisata['ekologi_dampak_cemar'].replace([[1, 2, 3], 0], [0, 1])
aspekDesaWisata.head(2)
```

```python
def scoring1(row):
    if row == 1:
        val=3
    else:
        val=0
    return val

def scoring2(row):
    if row == 1:
        val=3
    elif row == 2:
        val=2
    else:
        val=0
    return val
```

```python
aspekDesaWisata['ekonomi_hotel_score'] = aspekDesaWisata['ekonomi_hotel'].apply(scoring1)
aspekDesaWisata['ekonomi_produk_unggul_score'] = aspekDesaWisata['ekonomi_produk_unggul'].apply(scoring1)
aspekDesaWisata['ekologi_dampak_cemar_score'] = aspekDesaWisata['ekologi_dampak_cemar'].apply(scoring1)
aspekDesaWisata['ekonomi_status_bumdesa_score'] = aspekDesaWisata['ekonomi_status_bumdesa'].apply(scoring1)
aspekDesaWisata['ekonomi_bumdes_bid_pariwisata_score'] = aspekDesaWisata['ekonomi_bumdes_bid_pariwisata'].apply(scoring1)
aspekDesaWisata['ekonomi_pariwisata_wisata_desa_score'] = aspekDesaWisata['ekonomi_pariwisata_wisata_desa'].apply(scoring1)
aspekDesaWisata['ekonomi_pariwisata_agrowisata_score'] = aspekDesaWisata['ekonomi_pariwisata_agrowisata'].apply(scoring1)
aspekDesaWisata['ekonomi_pariwisata_wisata_alam_score'] = aspekDesaWisata['ekonomi_pariwisata_wisata_alam'].apply(scoring1)
aspekDesaWisata['ekonomi_pariwisata_transportasi_score'] = aspekDesaWisata['ekonomi_pariwisata_transportasi'].apply(scoring1)
aspekDesaWisata['ekologi_airdesa_score'] = aspekDesaWisata['ekologi_airdesa'].apply(scoring2)
aspekDesaWisata['ekonomi_transumum_score'] = aspekDesaWisata['ekonomi_transumum'].apply(scoring2)
aspekDesaWisata['ekonomi_jalan_desa_score'] = aspekDesaWisata['ekonomi_jalan_desa'].apply(scoring2)
aspekDesaWisata['ekonomi_kualitas_jalan_score'] = aspekDesaWisata['ekonomi_kualitas_jalan'].apply(scoring2)
aspekDesaWisata['sosial_bidang_pemukiman_sinyal_score'] = aspekDesaWisata['sosial_bidang_pemukiman_sinyal'].apply(scoring2)
aspekDesaWisata['total_score'] = aspekDesaWisata['ekonomi_hotel_score'] + aspekDesaWisata['ekonomi_produk_unggul_score'] + aspekDesaWisata['ekologi_dampak_cemar_score'] + aspekDesaWisata['ekonomi_status_bumdesa_score'] + aspekDesaWisata['ekonomi_bumdes_bid_pariwisata_score'] + aspekDesaWisata['ekonomi_pariwisata_wisata_desa_score'] + aspekDesaWisata['ekonomi_pariwisata_agrowisata_score'] + aspekDesaWisata['ekonomi_pariwisata_wisata_alam_score'] + aspekDesaWisata['ekonomi_pariwisata_transportasi_score'] + aspekDesaWisata['ekologi_airdesa_score'] + aspekDesaWisata['ekonomi_transumum_score'] + aspekDesaWisata['ekonomi_jalan_desa_score'] + aspekDesaWisata['ekonomi_kualitas_jalan_score'] + aspekDesaWisata['sosial_bidang_pemukiman_sinyal_score']
aspekDesaWisata = aspekDesaWisata.sort_values(by=['total_score'], ascending=False).reset_index(drop=True)

aspekDesaWisata.head(2)
```

Menurut (Yuniarti   et   al.,   2018) Indeks    kelayakan    dapat    dihitung    dengan  membandingkan nilai pada tiap kriteria dengan nilai maksimal pada tiap kriteria dalam persen. Kemudian   skor   yang   didapat   dari   setiap variabel akan menentukan  tingkat kelayakanya menggunakan rumus interval yaitu <br>



```python
aspekDesaWisata['indeks_kelayakan (%)'] = round((aspekDesaWisata['total_score'] / 42) * 100, 2)
aspekDesaWisata.head(2)
```

Menurut (Soekmadi  &  Kartodihardjo, 2010) untuk  menilai  indeks  kelayakan  terbagi menjadi 3 tingkatan yaitu : 
- Indeks  kelayakan  >  66.6%  :  Kawasan yang layak dikembangkan (Berpotensi Tinggi = 1)
- Indeks kelayakan 33.3% hingga 66.66% : Kawasan   belum   layak   dikembangkan (Berpotensi Sedang = 2) 
- Indeks  kelayakan  <  33.3%  :  Kawasan tidak layak dikembangkan (Berpotensi Rendah = 3)

```python
def kelas(row):
    if row >= 66.6:
        val=1
    elif row >= 33.3 and row <= 66.66:
        val=2
    else:
        val=3
    return val
```

```python
aspekDesaWisata['kelas'] = aspekDesaWisata['indeks_kelayakan (%)'].apply(kelas)
aspekDesaWisata.head(2)
```

```python
df_Final = pd.merge(df, aspekDesaWisata[['kemendagri_kode_kecamatan', 'kemendagri_kode_desa_kelurahan', 'kelas']],
                    on=['kemendagri_kode_kecamatan', 'kemendagri_kode_desa_kelurahan'], how='inner')
df_Final.head(2)
```
